### Aplicação dos Algoritmos Genéticos
Vamos utilizar as funções que definimos em um *script* de funções para facilitar o nosso trabalho.

```python
from pprint import pprint

from funcoes_liga_ternaria import populacao_liga_ternaria as cria_populacao
from funcoes_liga_ternaria import funcao_objetivo_pop_liga_ternaria as funcao_objetivo
from funcoes_liga_ternaria import selecao_torneio_max as funcao_selecao
from funcoes_liga_ternaria import cruzamento_liga_ternaria as funcao_cruzamento
from funcoes_liga_ternaria import mutacao_liga_ternaria as funcao_mutacao
```

Vamos detalhar o que cada uma dessas funções faz:
- `populacao_liga_ternaria`: Essa função cria a população para o nosso problema da liga ternária. O que ela faz é criar vários candidatos que seguem os critérios definidos no problema, como o peso total da liga ser 100g e os valores mínimos para cada elemento ser 5g. A estrutura de um candidato é uma lista na qual os três primeiros valores correspondem a massa daquele elemento na liga e os três últimos são os números inteiros que correspondem aos elementos no nosso dicionário. 

- `funcao_objetivo_pop_liga_ternaria`: A função objetivo do nosso problema é o valor da liga em dólares. Como queremos que esse valor seja o maior possível, temos um problema de maximização. Essa função é calculada multiplicando a massa de cada elemento pelo seu preço, e em seguida somando o preço de todos os elementos da liga.

- `selecao_torneio_max`: Usamos como operador de seleção o método de torneio para maximização, no qual ele irá pegar uma amostra de 3 indivíduos da nossa população e selecionar o que apresentar o melhor valor de *fitness*.

- `cruzamento_liga_ternaria`: Na nossa função de cruzamento, como o indivíduo é formado por duas partes que não podem se misturar, o ponto de corte do cruzamento deve ser exatamente no meio do indivíduo. Com isso, há a troca dos elementos e das massas para cada um dos filhos.

- `mutacao_liga_ternaria`: Para a nossa função de mutação, precisamos seguir a mesma lógica da operação de cruzamento, as duas metades do indivíduo são independentes uma da outra. Devido a isso, escolhemos criar um operador de mutação que sorteia se ele irá mutar as massas do indivíduo, ou algum de seus elementos. Caso ele mude as massas, todas elas são mudadas juntas, já que ao final elas precisam somar 100. Já se um elemento for mutado, eu sorteio um novo gene que ainda não esteja naquela liga.

Definidos os operadores e funções que iremos utilizar, vamos determinar os parâmetros do nosso algoritmo:

```python
TAMANHO_POPULACAO = 200
NUM_GERACOES = 10000
CHANCE_DE_CRUZAMENTO = 0.5
CHANCE_DE_MUTACAO = 0.05
TAMANHO_TORNEIO = 3

VALOR_MINIMO = 5
VALOR_MAXIMO = 90
```

Vamos analisar como é a nossa população:

```python
populacao = cria_populacao(TAMANHO_POPULACAO, elementos, VALOR_MINIMO, VALOR_MAXIMO)
pprint(populacao)
# funcao_objetivo(populacao)
```

Com tudo definido, podemos fazer a evolução do nosso algoritmo genético:

```python
hall_da_fama = []

for n in range(NUM_GERACOES):
    
    # Seleção
    fitness = funcao_objetivo(populacao, elementos)        
    selecionados = funcao_selecao(populacao, fitness, TAMANHO_TORNEIO)
    
    # Cruzamento
    proxima_geracao = []
    for pai, mae in zip(selecionados[::2], selecionados[1::2]):
        individuo1, individuo2 = funcao_cruzamento(pai, mae, CHANCE_DE_CRUZAMENTO)
        proxima_geracao.append(individuo1)
        proxima_geracao.append(individuo2)
    
    # Mutação
    funcao_mutacao(proxima_geracao, elementos, CHANCE_DE_MUTACAO, VALOR_MAXIMO, VALOR_MAXIMO)
    
    # Atualização do hall da fama
    fitness = funcao_objetivo(proxima_geracao, elementos)
        
    maior_fitness = max(fitness)
    indice = fitness.index(maior_fitness)
    hall_da_fama.append(proxima_geracao[indice])    
    
    # Encerramento
    populacao = proxima_geracao
```

Após toda a execução do algoritmo, temos o melhor indivíduo encontrado. Queremos saber qual seu preço e composição, que é o que temos abaixo:

```python
fitness = funcao_objetivo(hall_da_fama, elementos)
maior_fitness = max(fitness)
indice = fitness.index(maior_fitness)
melhor_individuo_observado = hall_da_fama[indice]

melhor_individuo_observado

print(f"Melhor indivíduo: {melhor_individuo_observado}")
print(f"Preço: ${maior_fitness:.2f}\n")
print("Composição:")

for i in range(3):
    print(f"{melhor_individuo_observado[i]}g de {elementos[melhor_individuo_observado[i+3]][0]}")
```

Pelo que já esperávamos, o melhor indivíduo realmente teria o Polônio como elemento mais presente, tendo em vista que ele é o mais caro, seguido do Actinio e do Califórnio. Ou seja, o algoritmo foi capaz de encontrar o melhor indivíduo possível para o problema.
